```python
import numpy as np
import matplotlib.pyplot as plt
```

## Case 1: $P(\tau) = \frac{c}{\mathcal{Z}} e^{-c(\tau-a)}$

### c > 0

```python
# Sample random variates
a = 0.5
b = 1.4
c = 3.14

Z = 1 - np.exp(-c*(b-a))
x = np.random.random(100000)
y = a - 1/c*np.log(1-Z*x)
```

```python
# Generate exact distribution
τ = np.linspace(a,b,1000)
P = (c/Z) * np.exp(-c*(τ-a))
```

```python
# Plot
plt.hist(y,bins=np.linspace(0,b,1000),density=True,label=r'$\rm{sampled}$');
plt.plot(τ,P,label=r'$\rm{exact}$',lw=2,color='k')
plt.xlabel(r'$\tau$')
plt.legend(frameon=False)
plt.savefig("case1_cpos.pdf",dpi=300)
```
